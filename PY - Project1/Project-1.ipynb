3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
'''
Выделяем признак опыта рабоыт в месяцах на основе признака 'Опыт работы' 
Пишем функцию для преобразования
Применяем к дата сету

Удлаяем признак 'Опыт работы' 

Отвечаем на вопросы
'''

def exp2months(x):
    '''
    Функция преобразования опыта выраженного в строковом формате в годах и месяцах в число месяцев
    '''

    if x is np.nan: # Обрабатываем пропуски в изначальном признаке
        return np.nan
    
    elif x == 'Не указано': #Обрабатываем отсутствующие данные в изначальном признаке
        return np.nan
    
    result = 0
    years = ['лет','год','года'] # Задаем возможные варианты написания лет
    months = ['месяц','месяца','месяцев'] # Задаем возможные варианты написания месяца
    
    data = x.split('  ')[0].split()[2:] # обрезаем инзачальную строку
    
    for i in range(1,len(data)):
        if data[i] in years:
            result = result + 12 * int(data[i-1])
        if data[i] in months:
            result = result + int(data[i-1])
    
    return result


df['Опыт работы (месяц)'] = df['Опыт работы'].apply(exp2months)
display(df['Опыт работы (месяц)'].median()) #выводим медиану опыта работы

df.drop('Опыт работы',axis = 1, inplace = True)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
'''
Из признака 'Город, переезд, командировки' выделяем 3 новых признака
Город выделяется при помощи split и маппинга с категориями

Для выделения Готовности к переезду используется функция readiness()

Для выделения признака готовности к командировкам используется сплит

Удаляем 'Город, переезд, командировки' признак

Отвечаем на вопросы
'''

display(df['Город, переезд, командировки'].head(100))
display(df['Город, переезд, командировки'][44740])
main_cities = ['Москва', 'Санкт-Петербург']
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
df['Город'] = df['Город, переезд, командировки'].apply(lambda x : x.split(' ,')[0] if x.split(' ,')[0] in main_cities else 'город-миллионник' if x.split(' ,')[0] in million_cities else 'другие') # Выделяем признак Город делая срез по шаблону и присваивая категорию

def readiness(x):
    '''
    Функция для выделения заданных фраз (ready, not_ready) и создание boolean признака на их основе
    '''
    ready = ['готов к переезду','готова к переезду','хочу переехать']
    not_ready = ['не готов к переезду', 'не готова к переезду']
    
    template = ' м.' # Определяем, шаблон, который будет использоваться для определения структуры строки изначального текста 
    data = x.split(' , ')
    if data[1][:3] == template:
        result = data[2] # При соотвествии шаблону, выбираем 3 элемент 
    else:
        result = data[1] # Если шаблон не встретился, выбираем 2 элемент
        
    result = result.split(' (')[0]
    
    if result in ready:
        return True
    elif result in not_ready:
        return False
    else:
        return np.nan
    
    
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(readiness) # Выделяем нужный признак
display(df['Готовность к переезду'].isnull().sum())


business_trip = ['готов к редким командировкам','готова к командировкам','готова к редким командировкам','готов к командировкам'] # Шаблон для признка коммандировки
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(lambda x : True if x.split(' , ')[-1] in business_trip else False) # берем последений элемент строки, заполняем boolean признаком, игнорируя пропуски

display(df['Готовность к командировкам'])
display(df.info())
display(df.nunique())


display(round(df.value_counts('Город',normalize = True)['Санкт-Петербург']*100)) # отвечаем про долю резюме из СПБ
display(df.value_counts(['Готовность к командировкам','Готовность к переезду'],normalize = True)) #Ищем долю готовых к переезду и командировкам

df.drop('Город, переезд, командировки', axis = 1, inplace = True)#ваш код здесь
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
'''
На основе знаения в полях Занятость и График создаем boolean признаки-маркеры
Использвем функцию transformation, которая позволяет определить наличе или отсутсие нужно признака

Отвечаем на вопросы, удаляем ставшими ненужными признаки

'''

def transformation(data, field2find):
    ''' 
    Функция возвращает True если запрашиваемый индекс находится в признаке
    '''
    for i in data.split(', '): 
        if i == field2find: # разбиваем признак на подстроки и проверяем значение в каждой подстроке
            return True
    return False


list_properties = ['Занятость','График'] #список значений по которым будем выделять признаки-маркеры
list_marker_properties = [] #создаем будующий список уникальных признаков-маркеров

'''
формируем уникальный список признаков маркеров для каждого искомого признака
'''
for i in list_properties:
    properties = set()
    for j in df[i].unique():
        for z in j.split(', '):
            properties.add(z)
    list_marker_properties.append(properties)

'''
В цикле перебираем и заполняем признаки маркерами
'''

for i in range(0,len(list_marker_properties)):
    for j in list_marker_properties[i]:
        df[j] = df[list_properties[i]].apply(transformation,args = ([j]))
        
        
display(df.head())

display(df[(df['проектная работа'] == True) & (df['волонтерство'] == True)]['Авто'].count()) # ищем ответ на вопрос Сколько людей ищут проектную работу и волонтёрство (в обоих столбцах стоит True)?
display(df[(df['вахтовый метод'] == True) & (df['гибкий график'] == True)]['Авто'].count())# ищем ответ на вопрос Сколько людей хотят работать вахтовым методом и с гибким графиком (в обоих столбцах стоит True)?
```

Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Пропорция - это число, за сколько единиц валюты указан курс в таблице с курсами. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.
Воспользуйтесь этой информацией в ваших преобразованиях.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
'''
Считываем новый data frame с курсами валют
Преобразовываем dataframe df, выделяем зарплату и валюту

Склеиваем 2 таблицы

Вычисляем желаемую зарплату в рублях по курсу 
Производим преобразования 

Отвечаем на вопросы

Удаляем ненужные признаки

'''


rates = pd.read_csv('data/ExchangeRates.csv') # Новый датафрейм с курсами валют
rates.drop(['per','time','vol'],axis = 1, inplace = True) # Отбрасываем ненужные поля
display(rates.head(20)) # Исследуем данные
display(rates.info())
display(rates[rates['currency'] == 'USD']['date']) # изучаем как огранизованы данные по 1 валюте
rates['date'] = rates['date'].astype('datetime64[ns]') # преобразовываем данные, по которым будет данные будут объединяться
display(rates.info())

''' Преобразования основного датафрейма df '''

df['date'] = df['Обновление резюме'].astype('datetime64[ns]').dt.date # Преобразуем данные в основном датафрейме и отбрасываем время
df['date'] = df['date'].astype('datetime64[ns]') 
display(df['date'].info())
display(df['date'])

df['salary_value'] = df['ЗП'].apply(lambda x : x.split(' ')[0]) # Выделяем целочисленный признак закрплаты
df['salary_value'] = df['salary_value'].astype('int64') 


currency_dict = {'грн.':'UAH','USD':'USD','бел.руб.':'BYN','KGS':'KGS', 'сум': 'UZS','AZN':'AZN','KZT':'KZT','руб.':'RUB','EUR':'EUR'}
df['currency'] = df['ЗП'].apply(lambda x: currency_dict[x.split(' ')[1]]) #Выделяем новый признак currency в формате ISO


'''Объединение таблиц'''

df_merge2 = pd.merge(df,rates, how='left',on = ['date','currency']) # Объединяем по 2 признакам

display(df_merge2.info()) # Проверяем осн информциию
display(df_merge2.isnull()) # Проверяем нули
display(df_merge2[df_merge2['currency'] == 'RUB']) # Подтвержаем, что nan у всех записей, указанных в RUB, т.к вспомогательная таблица не содержит обменного курса для RUB


df_merge2['close'].fillna(1, inplace = True) # заполняем пустые значения 1ми,
df_merge2['proportion'].fillna(1, inplace = True) # аналогично заполняем 1ми

df_merge2['ЗП (руб)'] = df_merge2['salary_value'] * df_merge2['close'] / df_merge2['proportion'] # считаем желаемый признак

display(round(df_merge2['ЗП (руб)'].median()/1000)) # счтает медианное значение по таблице

df_merge2.drop(['ЗП', 'Занятость','График','date','salary_value','close','currency','proportion'], axis = 1,inplace = True) # Удаляем ненужные признаки

df = df_merge2
display(df.info()) # производим сверку
```
